```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Read into the CSV file
csv_file = "Data_Source/USvideos.csv"
csv_df = pd.read_csv(csv_file)

# There was an error ParserError: Error tokenizing data. C error: Expected 11 fields in line 2401, saw 21
# We have to go into the csv file and actually fix the extra columns
csv_df.head(25)
```

```python
video_df = csv_df[["video_id", "title", "channel_title", "category_id", "views", "likes", "dislikes", "comment_total"]]
video_df.head(5)
```

```python
video_df['video_id'].value_counts()
```

```python
video_df.drop_duplicates("video_id",inplace=True)
```

```python
video_df.set_index("video_id", inplace=True)
```

```python
video_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/youtube_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
video_df.to_sql(name='videos', con=engine, if_exists='append', index=True)
```
